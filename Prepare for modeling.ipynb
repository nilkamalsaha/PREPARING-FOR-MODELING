```python
# outer join

happy_scores.merge(pop, how = 'outer', left_on = 'country_name', right_on = 'Country or territory')
```

```python
# right join

happy_scores.merge(pop, how = 'right', left_on = 'country_name', right_on = 'Country or territory')
```

```python
# 3. Creating a single table
```

```python
#connect to a sql database
import pandas as pd
import sqlite3

conn = sqlite3.connect('online_shop.db')
```

```python
#view the transactions table

transactions = pd.read_sql('Select * from transactions', conn)
transactions.head()
```

```python
#view the items table

items = pd.read_sql('Select * from items', conn)
items.head()
```

```python
# inner join

transactions.merge(items, how = 'inner', left_on = 'item_id', right_on = 'item_id')
```

```python
# outer join

transactions.merge(items, how = 'outer', left_on = 'item_id', right_on = 'item_id')
```

```python
# left join

df = transactions.merge(items, how = 'left', left_on = 'item_id', right_on = 'item_id')
df
```

```python
# right join

transactions.merge(items, how = 'right', left_on = 'item_id', right_on = 'item_id')
```

```python
# 4. Preparing Rows for modeling
```

```python
df
```

```python
#data types
df.dtypes
```

```python
# convert datatypes to datetime and numeric fields

df.purchase_date = pd.to_datetime(df.purchase_date)
df.price = pd.to_numeric(df.price.str.replace('$', ''))
df.rating = pd.to_numeric(df.rating)
```

```python
#checking data type again
df.dtypes
```

```python
#create a subset of april and may data
df_april_may = df[df.purchase_date.dt.month < 6]
df_april_may.head()
```

```python
#create a subset of june data
df_june = df[df.purchase_date.dt.month == 6]
df_june.head()
```

```python
# create a column of june dog food purchases

dog_food_rows = df_june[df_june.item_description == 'Dog Food']
```

```python
june_dog_food_purchases = (dog_food_rows
                         .groupby('customer')['item_id']
                         .count()
                         .rename('june_dog_food_purchases'))
june_dog_food_purchases
```

```python
# How much did each customer spend in april and may?

total_spend = df_april_may.groupby('customer')['price'].sum().rename('total_spend')
total_spend
```

```python
#each row represents a customer

model_df = (pd.concat([june_dog_food_purchases, total_spend], axis = 1)
            .fillna(0)
            .reset_index()
            .rename(columns = {'index':'customer'}))

model_df
```

```python
#Preparing Columns for modeling
# a. Text to numeric(Dummy Variables)
model_df.head()
```

```python
#goal: change the category column into numeric values
```

```python
df_april_may.head()
```

```python
# specify a column to get dummy variables
category_dummies = pd.get_dummies(df_april_may.category)
category_dummies
```

```python
#combine with customer data
pd.concat([df_april_may.customer, category_dummies], axis = 1)
```

```python
# groupby by customer so each row is a single customer

categories = pd.concat([df_april_may.customer, category_dummies], axis=1).groupby('customer').sum().reset_index()
categories.head()
```

```python
#add categories to the model dataframe
model_df_categories = model_df.merge(categories, how='left', on ='customer')
model_df_categories.head()
```

```python
# goal: add on a new column based on date (number of days since last purchase)
```

```python
# find the last purchase date
last_purchase = df_april_may.groupby('customer')['purchase_date'].max()
last_purchase.head()
```

```python
# note down today's date

today = pd.Series(pd.to_datetime('2023-06-01'), index = last_purchase.index)
today.head()
```

```python
# find the number of days since the last purchase

days_between = (today - last_purchase).dt.days.rename('days_between')
days_between.head()
```

```python
# add the column to the model_df_categories dataframe
model_df_categories = model_df_categories.merge(days_between.reset_index(), how ='left', on='customer')
model_df_categories.head()
```
